```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn.cluster import KMeans 
```

```python
data = pd.read_csv("Mall_Customers.csv")
data.head()
```

```python
data['Genre'] = np.where(data['Genre']== 'Male', 1,0)
```

```python
data.head()
```

```python
X = data.iloc[ :, 3:5].values
```

```python
X
```

```python
WCSS_LIST  = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++',random_state = 42)
    kmeans.fit(X)
    WCSS_LIST.append(kmeans.inertia_)
plt.plot(range(1, 11), WCSS_LIST)
plt.title('Le graphique de la methode Elow')
plt.xlabel('Le graphique de la methode Elow')
plt.ylabel('WCSS_LIST')
plt.show
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)

y_predict = kmeans.fit_predict(X)
```

```python
plt.scatter(X[y_predict == 0,0], X[y_predict == 0,1], s = 100, c = 'blue', label = 'Goupe 1')
plt.scatter(X[y_predict == 1,0], X[y_predict == 1,1], s = 100, c = 'green', label = 'Goupe 2')
plt.scatter(X[y_predict == 2,0], X[y_predict == 2,1], s = 100, c = 'red', label = 'Goupe 3')
plt.scatter(X[y_predict == 3,0], X[y_predict == 3,1], s = 100, c = 'cyan', label = 'Goupe 4')
plt.scatter(X[y_predict == 4,0], X[y_predict == 4,1], s = 100, c = 'magenta', label = 'Goupe 5')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 300, c= 'yellow', label = 'Centroide')
plt.title('Groupe de client')
plt.xlabel('Revenu annuel (k$)')
plt.ylabel('Score de depenses (1-100)')
plt.legend()
plt.show()
```
